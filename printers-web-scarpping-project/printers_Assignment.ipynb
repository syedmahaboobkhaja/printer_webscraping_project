## Q6. How many products use specific keywords in their titles or descriptions (e.g., “organic”, “wireless”, “gaming”), and what fraction of the total category is this?


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}

data = []

for page in range(1, 11):
    url = f"https://www.flipkart.com/search?q=printers&page={page}"
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    products = soup.find_all('div', {'data-id': True})
    
    for product in products:
        # METHOD 1: Try 'title' attribute
        name_tag = product.find('a', class_='GnxRXv')
        name = None
        
        if name_tag and name_tag.get('title'):
            name = name_tag.get('title')
        
        # METHOD 2: Try img 'alt' attribute
        if not name or name == 'N/A':
            img_tag = product.find('img', class_='UCc1lI')
            if img_tag and img_tag.get('alt'):
                name = img_tag.get('alt')
        
        # METHOD 3: Try text from link
        if not name or name == 'N/A':
            if name_tag:
                text = name_tag.get_text(strip=True)
                # Extract name before rating (e.g., before "4.3(")
                if text:
                    name = text.split('(')[0].replace('₹', '').strip()
        
        name = name if name else 'N/A'
        brand = name.split()[0] if name != 'N/A' else 'N/A'
        
        # Get Rating
        rating_tag = product.find('div', class_='a7saXW')
        rating = 'N/A'
        if rating_tag:
            rating_text = rating_tag.text.strip()
            rating = rating_text.split('(')[0] if '(' in rating_text else rating_text
        
        # Get Price
        price = 'N/A'
        price_links = product.find_all('a')
        for link in price_links:
            text = link.text.strip()
            if '₹' in text:
                price = text.split('₹')[1].split('₹')[0] if text.count('₹') > 1 else text
                price = '₹' + price
                break
        
        data.append({
            'Name': name,
            'Brand': brand,
            'Rating': rating,
            'Price': price
        })
    
    print(f"Page {page}: {len([d for d in data[-40:] if d['Name'] != 'N/A'])} names found")
    time.sleep(2)

df = pd.DataFrame(data)
df.to_csv('Flipkart_Printers_Data.csv', index=False)

print(f"\n✅ Total: {len(df)} products!")
print(f"✅ With Names: {len(df[df['Name'] != 'N/A'])} products")
print("\nFirst 5:")
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV
df = pd.read_csv('Flipkart_Printers_Data.csv')

# Keywords to search (FOR PRINTERS!)
keywords = ['wireless', 'laser', 'inkjet', 'multifunction', 'color']

total_products = len(df)
keyword_counts = {key: 0 for key in keywords}

# Search in NAME column
for idx, row in df.iterrows():
    title_text = str(row['Name']).lower()
    
    for key in keywords:
        if key in title_text:
            keyword_counts[key] += 1

# Create DataFrame
keyword_data = []
for key, count in keyword_counts.items():
    fraction = (count / total_products) * 100 if total_products > 0 else 0
    keyword_data.append({
        "keyword": key,
        "product_count": count,
        "percentage_of_category": fraction
    })

keyword_df = pd.DataFrame(keyword_data)
print(keyword_df)

# Plot
plt.figure()
plt.bar(keyword_df["keyword"], keyword_df["product_count"])
plt.xlabel("Keyword")
plt.ylabel("Number of Products")
plt.title("Keyword Presence in Printer Titles (First 10 Pages)")
plt.show()
```

## Q7. Among top N search results for a keyword, how many products are “sponsored” or “ads” vs. organic results, if the site marks them differently?


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}

N = 50  # Top N products to analyze

url = "https://www.flipkart.com/search?q=printers"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

products = soup.find_all('div', {'data-id': True})[:N]

sponsored_count = 0
organic_count = 0

for product in products:
    # Check if product is sponsored
    # Flipkart marks sponsored products with specific text or class
    is_sponsored = False
    
    # Method 1: Check for "Sponsored" text
    sponsored_tag = product.find(text=lambda t: t and 'sponsored' in t.lower())
    if sponsored_tag:
        is_sponsored = True
    
    # Method 2: Check for ad-related classes
    ad_class = product.find('div', class_=lambda c: c and 'ad' in str(c).lower())
    if ad_class:
        is_sponsored = True
    
    if is_sponsored:
        sponsored_count += 1
    else:
        organic_count += 1

# Calculate percentages
sponsored_pct = (sponsored_count / N) * 100 if N > 0 else 0
organic_pct = (organic_count / N) * 100 if N > 0 else 0

print(f"Analysis of Top {N} Search Results for 'printers':\n")
print(f"Sponsored/Ads: {sponsored_count} ({sponsored_pct:.2f}%)")
print(f"Organic Results: {organic_count} ({organic_pct:.2f}%)")
print(f"\nTotal: {N} products")

# Create visualization
import matplotlib.pyplot as plt

labels = ['Sponsored', 'Organic']
counts = [sponsored_count, organic_count]
colors = ['#FF6B6B', '#4ECDC4']

plt.figure(figsize=(8, 6))
plt.bar(labels, counts, color=colors)
plt.xlabel('Result Type')
plt.ylabel('Number of Products')
plt.title(f'Sponsored vs Organic Results (Top {N} Products)')

# Add count labels on bars
for i, count in enumerate(counts):
    plt.text(i, count + 1, str(count), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Also create pie chart
plt.figure(figsize=(8, 6))
plt.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title(f'Sponsored vs Organic Distribution (Top {N} Products)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

After analyzing the top 50 search results for printers, I observed that no products had any visible "Sponsored" or "Ad" labels. All 50 products appeared as regular search results.

This could mean that Flipkart doesn't show sponsored products in the printer category, or the sponsored tags are not easily identifiable through basic web scraping. It's possible that ads are displayed in other formats like banners or separate sections which weren't captured in this analysis.

This gap represents an opportunity for new brands to enter the market. By offering good quality printers at competitive prices, new companies can attract budget-conscious customers who want reliable products without paying premium prices.
