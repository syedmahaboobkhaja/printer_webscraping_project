```python
import re
import numpy as np
```

```python
#for multiple classes
soup.find_all("a",class_="pIpigb")
```

```python
#to remove repeated values 
brand=[]
for i in soup.find_all("a",class_="pIpigb"):
    print(i.text)
    regex=re.findall("^[a-zA-Z]+",i.text)
    print(regex)
    if regex:
        brand.append(regex[0])
    else:
        brand.append(np.nan)
```

```python
rating=[]
for i in soup.find_all("div",class_="MKiFS6"):
    print(i.text)
    rating.append(i.text)
```

```python
price=[]
for i in soup("div",class_="hZ3P6w"):
    print(i.text)
    price.append(i.text)
```
